```python
def addnewstudent():
    for t in range(3):           
        name = input("Name : ")
        id = abs(hash(name)%(10**4))
        if isstudent(id):
            print("Student already present in list..")
        else:   
            c = input("""
                p for Phd Student
                m for Msc Student

            Enter Course : 
              """)
            if c=='p':
                std.append(PhdStudent(name, id, int(input("M1 : ")), int(input("M2 : ")),
                      int(input("M3 : ")), input("Thesis : "), int(input("Grade : "))))
                print("Student data added successfully......")
                return True
            elif c=='m':
                std.append(MScStudent(name, id, int(input("M1 : ")), int(input("M2 : ")),
                      int(input("M3 : ")), input("Special Subject : "), int(input("M4 : "))))
                print("Student data added successfully......")
                return True
            else:
                print("wring course try next time....")
                return False
    print("Your completed your 3 term...")
```

```python
def subjectmodify():
    name = input("Enter name of Student : ")
    for s in std:
        if s.get_Name()==name:
            s.set_subject(input("Enter new subject : "))
            print("Subject changed successfully...")
            return 0
    print(f"Studen {name} not found")
```
